```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Load the cleaned dataset
file_path = "colleges.csv"
colleges_df = pd.read_csv(file_path, index_col=0)
colleges_df.head()
```

```python
colleges_df.hvplot.scatter(x="enrolled_total", y="average_tuition")
```

```python
colleges_df.describe()
```

```python
colleges_df
```

```python
colleges_train, colleges_test = train_test_split(colleges_df,random_state=1)
```

```python
colleges_train.shape
```

```python
colleges_test.shape
```

```python
def get_splitclusters(k, data):
    # Create a copy of the DataFrame
    data = data.copy()

    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Fit the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["splitcluster"] = model.labels_

    return data
```

```python
four_clusters_train=get_splitclusters(4, colleges_train)
four_clusters_train.head()
```

```python
four_clusters_test=get_splitclusters(4, colleges_test)
four_clusters_test.head()
```

```python
colleges_split = four_clusters_train.append(four_clusters_test)
colleges_split.head()
```

```python
colleges_split.to_csv('colleges_split.csv')
```

```python
colleges_train.to_csv('colleges_train.csv')
```

```python
def get_fullclusters(k, data):
    # Create a copy of the DataFrame
    data = data.copy()

    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Fit the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["fullcluster"] = model.labels_

    return data
```

```python
colleges_4clusters=get_fullclusters(4, colleges_df)
colleges_4clusters.head()
```

```python
colleges_4clusters.hvplot.scatter(x="total_enrollment", y="average_tuition", by="fullcluster")
```

```python
colleges_split.hvplot.scatter(x="total_enrollment", y="average_tuition", by="splitcluster")
```
